# Exploratory Data Analysis with Python

### Import Python libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

### Loading Data Set

Printing the first 10 rows

```python
df = pd.read_csv("CarRental.csv")
df.head(10)
```

Printing the last 5 rows

```python
df.tail(5)
```

### Checking the types of data

```python
df.dtypes
```

### Removing Irrelevant column

```python
df = df.drop(['location.country', 'owner.id'], axis=1)
df.head(5)
```

### Rename Columns

```python
df = df.rename(columns={"fuelType": "Fuel", "renterTripsTaken": "C_Trips", "reviweCount": "Review", "location.city": "City","location.latitude": "Latitude", "location.longitude": "Longitude", "location.state": "State" })
df.head(5)
```

```python
df = df.rename(columns={"fuelType": "Fuel", "renterTripsTaken": "C_Trips", "reviweCount": "Review", "location.city": "City","location.latitude": "Latitude", "location.longitude": "Longitude", "location.state": "State", "rate.daily": "MilesD", "vehicle.make": "Maker", "vehicle.model": "Model", "vehicle.type": "V_Body", "vehicle.year": "year"  })
df.head(5)
```

## Duplicate observations (Colunms)

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
df.count()
```

### Dropping the missing or null values.

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```
